### Cretating the transform function

```python
# Here Compose is used to stack all the transforms as a list.
# first transformation we ned as horizontalflip. it is similer to flipping your palm upside down, fliping.
# 


transform = transforms.Compose([ 
    transforms.RandomHorizontalFlip(0.9), # to flip the image like we flip the palm.
    transforms.RandomRotation(90), # rotatwe by 90 degree,
    transforms.Resize((250, 250)), # we need to have all the images one size so that our model can be trained effeciently.
    transforms.ToTensor() # in the molde PIL Image can not be fed so we need to convert it in to the trnsor
    
    
    
])
```
